## Encontrar todas as ocorencias de um caracter ou string

```python
re.findall(r'K3241',texto)
```

## Disjunção

```python
re.findall(r'K[0123456789]',texto)
```

```python
re.findall(r'K[0-9][0-9][0-9][0-9]',texto)
```

## Negação
* Exemplo emcontrar todas as ocorrências que não são números

```python
re.findall(r'[^0-9]',texto)
```

## Múltiplos Padrões de Busca
* Exemplo, encontrar todas as ocorrências  'k' e '0'

```python
re.findall(r'k|0', texto)
```

## Marcador de início do texto
* Exemplo: Todas as ocorrências de 'L' no icício do texto

```python
re.findall(r'^L', texto)
```

## Marcador de final do texto
* Exemplo: Todas as ocorrências de 'f' no final do texto

```python
re.findall(r'f$', texto)
```
