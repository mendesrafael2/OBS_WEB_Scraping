Atividade:

* Encontre o padrão para o download das tabelas PERTINENTES 
    * Empresa: Os nomes dos arquivos iniciam com K3241
    * Sócio: Os nomes dos arquivos iniciam com K3241
    * Estabelecimento: Os nomes dos arquivos iniciam com K3241  
* Use os meios necessários para refazer a lista contendo somente os links pertinentes.

```python
# Abrir URLs
import requests

# Navegar em tags HTML
from bs4 import BeautifulSoup 

import pandas as pd

# Trabalhar com expressões regulares
import re
```

## Acessar o site mencionado e fazer uma raspagem dos links contidos nele

```python
r = requests.get('http://200.152.38.155/CNPJ/')
soup = BeautifulSoup(r.text,'html.parser')

# soup.find_all('a') -> Encontra os dados que utilizam a tag <a>
links = soup.find_all('a')
```

```python
lista_links_pertinentes = []

for link in links:
    link_get = link.get('href')
    
    # Expressão regular utilizada
    link_filtrado = re.findall(r'K3241.K03200Y[0-9].D11113.{8}[A-Z].zip',link_get)
    
    if len(link_filtrado) != 0:
        lista_links_pertinentes.append(link_filtrado)
```

##  Transforme a lista em um DataFrame de coluna única

```python
df_links_pertinentes = pd.DataFrame(data=lista_links_pertinentes, columns=['Link'])
df_links_pertinentes
```

## Salve o DataFrame em um csv

```python
df_links_pertinentes.to_csv('links_pertinentes.csv',index=False)
```
